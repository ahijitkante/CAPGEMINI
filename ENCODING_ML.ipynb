```python
#label encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['salary_encoded'] = le.fit_transform(df['salary'])

print(df[['salary', 'salary_encoded']].head())
```

```python
#one hot encoding

import pandas as pd
from sklearn.preprocessing import OneHotEncoder
df = pd.read_csv("https://raw.githubusercontent.com/Taj1920/Data-Analysis-Datasets/refs/heads/main/Datasets-Practice/hr_data.csv")


ohe = OneHotEncoder(sparse_output=False)

# Fit and transform
encoded_data = ohe.fit_transform(df[['department']])

# Column names
encoded_columns = ohe.get_feature_names_out(['department'])

# Convert to dataframe
encoded_df = pd.DataFrame(encoded_data, columns=encoded_columns)

# Concatenate
df = pd.concat([df, encoded_df], axis=1)

# Drop original column
df.drop('department', axis=1, inplace=True)

print(df.head())
```

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/Taj1920/Data-Analysis-Datasets/refs/heads/main/Datasets-Practice/hr_data.csv")

# Create encoder
oe = OrdinalEncoder(categories=[['low', 'medium', 'high']])

# Fit and transform salary column
encoded_data = oe.fit_transform(df[['salary']])

# Replace original column
df['salary'] = encoded_data

print(df.head())
```
