```python
import pandas as pd
from sklearn.preprocessing import Normalizer

# Dataset
data = {
    "Height": [162, 171, 133, 165],
    "Weight": [63, 60, 50, 62],
    "Speed":  [5, 3, 7, 2]
}

df = pd.DataFrame(data, index=["S1", "S2", "S3", "S4"])
df

nom=Normalizer()

data=nom.fit_transform(df)
data

df=pd.DataFrame(data,columns=df.columns)
df
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Dataset
data = {
    "Height": [162, 171, 133, 165],
    "Weight": [63, 60, 50, 62],
    "Speed":  [5, 3, 7, 2]
}

df = pd.DataFrame(data, index=["S1", "S2", "S3", "S4"])
df

nom=StandardScaler()

data=nom.fit_transform(df)
data

df=pd.DataFrame(data,columns=df.columns)
df
```
