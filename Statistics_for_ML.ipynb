```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/winequalityN.csv')
```

```python
data
```

```python
pop=data
```

```python
pop=data['alcohol']
```

```python
sample_size=500
sample=data['alcohol'].sample(sample_size)
sample_mean=sample.mean()
pop_mean=pop.mean()
error=pop_mean-sample_mean
error
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(range(1,sample_size+1),sample)
```

```python
grouped = data.groupby('type')

white_wine = grouped.get_group('white')
red_wine = grouped.get_group('red')

pop_white=white_wine['alcohol']
sample_size_white=50
sample_white=pop_white.sample(sample_size_white)
sample_white_mean=sample_white.mean()
pop_white_mean=pop_white.mean()
error=pop_white_mean-sample_white_mean
error
```

```python
pop_red = red_wine['alcohol']
sample_size_red = 50
sample_red = pop_red.sample(sample_size_red)
sample_red_mean = sample_red.mean()
pop_red_mean = pop_red.mean()
error_red = pop_red_mean - sample_red_mean

error_red
```

```python
plt.scatter(range(1,sample_size_white+1),sample_white)
```

```python
plt.scatter(range(1, sample_size_red + 1), sample_red)

plt.axhline(pop_red_mean, color='red', label='Red Population Mean')
plt.axhline(sample_red_mean, color='blue', label='Red Sample Mean')

plt.xlabel("Sample Index")
plt.ylabel("Alcohol %")
plt.title("Red Wine Alcohol Sampling")
plt.legend()

plt.show()
```

```python
car_data=pd.read_csv('Automobile.csv')
car_data
```
